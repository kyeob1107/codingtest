```python
from itertools import combinations
def combinations2(iterable, r):
    # iterable을 리스트로 변환
    pool = tuple(iterable)
    n = len(pool)

    # 조합의 길이가 전체 길이보다 크면 빈 리스트 반환
    if r > n:
        return

    indices = list(range(r))
    yield tuple(pool[i] for i in indices)

    while True:
        for i in reversed(range(r)):
            if indices[i] != i + n - r:
                break
        else:
            return
        indices[i] += 1
        for j in range(i + 1, r):
            indices[j] = indices[j - 1] + 1
        yield tuple(pool[i] for i in indices)

# 예제 사용법
for combo in combinations2('ABCDE', 4):
    print(combo)
print("====================================")
for combo in combinations('ABCDE', 4):
    print(combo)
```

```python
def combinations(iterable, r=None):
    # iterable을 리스트로 변환
    pool = tuple(iterable)
    n = len(pool)
    
    # r이 None이면 모든 길이의 조합을 생성
    if r is None:
        for r in range(1, n + 1):
            yield from combinations(pool, r)
        return
    
    # 조합의 길이가 전체 길이보다 크면 빈 리스트 반환
    if r > n:
        return

    indices = list(range(r))
    yield tuple(pool[i] for i in indices)

    while True:
        for i in reversed(range(r)):
            if indices[i] != i + n - r:
                break
        else:
            return
        indices[i] += 1
        for j in range(i + 1, r):
            indices[j] = indices[j - 1] + 1
        yield tuple(pool[i] for i in indices)

# 예제 사용법
for combo in combinations('ABCD'):
    print(combo)
```

```python
from itertools import combinations, permutations
def permutations2(iterable, r=None):
    pool = tuple(iterable)
    n = len(pool)
    r = n if r is None else r
    
    if r > n:
        return
    
    indices = list(range(n))
    cycles = list(range(n, n-r, -1))
    
    yield tuple(pool[i] for i in indices[:r])
    
    while n:
        for i in reversed(range(r)):
            cycles[i] -= 1
            if cycles[i] == 0:
                indices[i:] = indices[i+1:] + indices[i:i+1]
                cycles[i] = n - i
            else:
                j = cycles[i]
                indices[i], indices[-j] = indices[-j], indices[i]
                yield tuple(pool[i] for i in indices[:r])
                break
        else:
            return

# 예제 사용법
for perm in permutations2('ABCD'):
    print(perm)

# r을 명시적으로 지정할 수도 있습니다
for perm in permutations2('ABCD', 2):
    print(perm)
    
for perm in permutations('ABCD'):
    print(perm)
```
