```python
# 잘 이해는 안감 초기값 잘못되있어서 데이터 추가후 틀렸다고 나오는 듯해서 그부분 수정하니 맞게나오는 거 같긴한데
import sys
# input = sys.stdin.readline
input = open("../../input.txt", "rt").readline

N = int(input())

nums = list(map(int, input().split()))

op_count = list(map(int, input().split()))

op_func = [
    lambda a, b: a+b,
    lambda a, b: a-b,
    lambda a, b: a*b,
    lambda a, b: int(a/b)
]

answer = [-10**9, 10**9]


def select(op_count, toPick, length, N):
    global answer

    if toPick == 0:
        answer[0] = max(answer[0], N)
        answer[1] = min(answer[1], N)
        return

    for i in range(4):
        if op_count[i] != 0:
            op_count[i] -= 1

            select(op_count, toPick-1, length,
                   op_func[i](N, nums[length-toPick+1]))

            op_count[i] += 1


select(op_count, sum(op_count), sum(op_count), nums[0])

print(answer[0])
print(answer[1])
```

```python
# 흠.. 어렵다
import sys
si = sys.stdin.readline
N = int(si())
A = list(map(int, si().split()))
op = list(map(int, si().split()))
M, m = -10**9, 10**9
target = 0
k = [11**i for i in range(11)]
for i, v in enumerate(op):
    target += op[i] * k[i]
memo = {}


def dfs(depth, bit, v):
    if bit == target:
        global M, m
        M = max(M, v)
        m = min(m, v)
        return
    if bit in memo:
        a, b = memo[bit]
        if a <= v <= b:
            return
        if v < a:
            memo[bit][0] = v
        else:
            memo[bit][1] = v
    else:
        memo[bit] = [v, v]

    for _ in range(4):
        if op[_]:
            op[_] -= 1
            if _ == 0:
                dfs(depth+1, bit+k[_], v+A[depth])
            elif _ == 1:
                dfs(depth+1, bit+k[_], v-A[depth])
            elif _ == 2:
                dfs(depth+1, bit+k[_], v*A[depth])
            else:
                if v < 0:
                    dfs(depth+1, bit+k[_], -(-v // A[depth]))
                else:
                    dfs(depth+1, bit+k[_], v//A[depth])
            op[_] += 1


dfs(1, 0, A[0])
print(M)
print(m)
```

```python
# 참고 레퍼런스2 - 제출하니 시간초과 안뜨는 경우는 틀렸습니다로 나옴
# 순열 (Python3 시간초과 / PyPy3는 통과)
import sys
from itertools import permutations

# input = sys.stdin.readline
input = open("../../input.txt", "rt").readline
N = int(input())
num = list(map(int, input().split()))
op_num = list(map(int, input().split()))  # +, -, *, /
op_list = ['+', '-', '*', '/']
op = []

for k in range(len(op_num)):
    for i in range(op_num[k]):
        op.append(op_list[k])

maximum = -1e9
minimum = 1e9

def solve():
    global maximum, minimum
    for case in permutations(op, N - 1):
        total = num[0]
        for r in range(1, N):
            if case[r - 1] == '+':
                total += num[r]
            elif case[r - 1] == '-':
                total -= num[r]
            elif case[r - 1] == '*':
                total *= num[r]
            elif case[r - 1] == '/':
                total = int(total / num[r])
	
        if total > maximum:
            maximum = total
        if total < minimum:
            minimum = total

solve()
print(maximum)
print(minimum)
```

```python
# 내가 처음 생각했던 순열 조합 짜는 것?
import sys
# input = sys.stdin.readline
input = open("../../input.txt", "rt").readline
N=int(input())
A=list(map(int,input().split()))
p,m,t,d=map(int,input().split())

l=[] 
pmtd=[] #연산자 순열 리스트

def bt():
    global N
    global p
    global m
    global t
    global d
    if len(pmtd)==N-1:
        l.append(pmtd.copy())# 값 추가
    else:
        for i in [(p,0),(m,1),(t,2),(d,3)]:
            if not i[0]:
                continue
            if i[1]==0:
                pmtd.append('p')
                p-=1
            elif i[1]==1:
                pmtd.append('m')
                m-=1
            elif i[1]==2:
                pmtd.append('t')
                t-=1
            else:
                pmtd.append('d')
                d-=1
            bt()
            pmtd.pop()
            if i[1]==0:
                p+=1
            elif i[1]==1:
                m+=1
            elif i[1]==2:
                t+=1
            else:
                d+=1

bt()
print(l)
```
