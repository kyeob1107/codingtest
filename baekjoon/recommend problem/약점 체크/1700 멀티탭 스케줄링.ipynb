```python
# 대충 후의 N칸 까지 나타나지 않는 것중에 사용빈도 제일 적게 남은 것 찾는 방식 - 거의 대부분 되긴하는데 틀렸다고 나오는데 반례 못찾겠음
import sys
# input = sys.stdin.readline
input = open("../../input.txt", "rt").readline
N, K = map(int, input().split())
use_history = list(map(int, input().split()))
# use_history.reverse() # 이게 원래
use_cnt = {}
for u in use_history:
    if u in use_cnt:
        use_cnt[u] += 1
    else:
        use_cnt[u] = 1
print(use_cnt)
if len(use_cnt.keys()) <= N:
    print(0)
else:
    use_history.reverse()        
    res = 0
    multitab = [] #[0] * N
    for i in range(K):
        check_list = set(use_history[-1*N:])
        use = use_history.pop()
        use_cnt[use] -= 1
        if use not in multitab:
            if len(multitab) < N:
                multitab.append(use)
            else:
                if all(m not in check_list for m in multitab):
                    min_use = min((k for k in use_cnt if k in multitab), key=use_cnt.get, default=None)
                else:
                    min_use = min((k for k in use_cnt if k not in check_list and k in multitab), key=use_cnt.get, default=None)
                multitab.remove(min_use)
                multitab.append(use)
                res += 1
        # print(multitab)
    print(res)
```

```python
import sys
# input = sys.stdin.readline
input = open("../../input.txt", "rt").readline
N, K = map(int, input().split())
use_history = list(map(int, input().split()))
use_order = {}
for i, u in enumerate(use_history):
    if u in use_order:
        use_order[u].append(i)
    else:
        use_order[u] = [i]

if len(use_order.keys()) <= N:
    print(0)
else:
    use_history.reverse()
    res = 0
    multitab = [] #[0] * N
    
    for i in range(K):
        use = use_history.pop()
        use_order[use].remove(i)
        if use not in multitab:
            if len(multitab) < N:
                multitab.append(use)
            else:
                next_order = [[u, use_order[u][0]] if len(use_order[u]) > 0 else [u, K+1] for u in multitab]
                later_use = max(next_order, key=lambda x: x[1])[0]
                multitab.remove(later_use)
                multitab.append(use)
                res += 1
    print(res)
```

```python
# 뤼튼이 개선해준 코드 1
import sys
# input = sys.stdin.readline
input = open("../../input.txt", "rt").readline
N, K = map(int, input().split())
use_history = list(map(int, input().split()))

# 각 기기의 사용 순서를 저장
use_order = {}
for i, u in enumerate(use_history):
    if u not in use_order:
        use_order[u] = []
    use_order[u].append(i)

# 멀티탭에 꽂을 필요가 없는 경우
if len(use_order) <= N:
    print(0)
    sys.exit()

res = 0
multitab = []

for i in range(K):
    use = use_history[i]
    use_order[use].pop(0)

    if use in multitab:
        continue

    if len(multitab) < N:
        multitab.append(use)
    else:
        # 다음에 사용할 기기 중 가장 나중에 사용할 기기를 찾음
        farthest_use = -1
        device_to_remove = -1

        for device in multitab:
            if use_order[device]:
                next_use = use_order[device][0]
            else:
                next_use = K + 1

            if next_use > farthest_use:
                farthest_use = next_use
                device_to_remove = device

        multitab.remove(device_to_remove)
        multitab.append(use)
        res += 1

print(res)
```

```python
# 1등 코드
N, K = map(int, input().split())
if N >= K :
    print(0)
    exit()

elec_list = list(map(int, input().split()))

plug = set()
cnt = 0

def find_latest(idx) :
    result = 0
    max_idx = -1
    for num in plug :
        try :
            num_idx = elec_list[idx:].index(num)
        except :
            num_idx = K
        if max_idx < num_idx :
            result, max_idx = num, num_idx
    return result
  
for idx, num in enumerate(elec_list) :
    plug.add(num)
    if len(plug) > N :
        cnt += 1
        latest_used = find_latest(idx)
        plug.discard(latest_used)

print(cnt)
```
