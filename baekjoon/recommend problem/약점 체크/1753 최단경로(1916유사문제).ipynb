```python
import heapq
import sys
# input = sys.stdin.readline
input = open("../../input.txt", "rt").readline
V, E = map(int, input().split()) # 정점, 간선
start = int(input().strip())
graph = [{} for _ in range(V + 1)]
for _ in range(E):
    u, v, w = map(int, input().split())
    if v in graph[u]:
        graph[u][v] = min(graph[u][v], w)
    else:
        graph[u][v] = w

init = float('inf')

def dijkstra(graph, start):
    distances = [init] * (V + 1)
    distances[start] = 0
    queue = [(0, start)]
    
    while queue:
        # print(queue) #
        current_distance, current_node = heapq.heappop(queue)
        # print(distances) #
        if current_distance > distances[current_node]:
            continue
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(queue, (distance, neighbor))
    
    return distances
# print(graph) #
for i, distance in enumerate(dijkstra(graph, start)):
    if i == 0:
        continue
    if distance == init:
        print("INF")
    else:
        print(distance)
```

```python
# pypy3 1등 코드
import sys,io,os,heapq
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline
output = io.StringIO()
N,E = map(int,input().split())
K = int(input())
G = {i:[] for i in range(1,N+1)}
for _ in range(E):
  u,v,w = map(int,input().split())
  G[u].append(w<<16|v)

D=[sys.maxsize]*(N+1)
D[K]=0
Q=[K]
while Q:
  n = heapq.heappop(Q)
  dist = n>>16
  v = n&0xffff
  if(dist > D[v]):
    continue
  for next_n in G[v]:
    next_dist = next_n>>16
    next_v = next_n&0xffff
    distance = dist + next_dist
    if(distance < D[next_v]):
      D[next_v] = distance
      heapq.heappush(Q,distance<<16|next_v)
for x in D[1:]:
  if x!=sys.maxsize:
    output.write(f"{x}\n")
  else:
    output.write("INF\n")
sys.stdout.write(output.getvalue())
```

```python
# 파이썬 1등 코드
import sys
import heapq as hq

def rL():
    return sys.stdin.readline().rstrip()

def dijikstra(graph, n, st):
    mintable = [sys.maxsize]*(n+1)
    mintable[st] = 0
    heap = [st]
    
    while heap:
        p = hq.heappop(heap)
        w = p>>16
        v = p&0xffff
        
        if w > mintable[v]:
            continue
        
        for i in graph[v]:
            nw = i>>16
            nv = i&0xffff
            nxdis = w+nw
            if nxdis < mintable[nv]:
                mintable[nv] = nxdis
                hq.heappush(heap, nxdis<<16|nv)
                       
    return mintable

def main():
    n,e = map(int,rL().split())
    graph = [[] for _ in range(n+1)]
    st = int(rL())
    
    for i in range(e):
        u,v,w = map(int,rL().split())
        graph[u].append(w<<16|v)
    
    mintable = dijikstra(graph,n,st)

    for i in mintable[1::]:
        if i != sys.maxsize:
            sys.stdout.write("%d\n"%i)
        else:
            sys.stdout.write("INF\n")
            

if __name__ == "__main__":
    main()
```

```python
# 파이썬 2등 코드
import sys
import heapq

def dijkstra(V, graph, start):
    # 무한대를 의미하는 값
    INF = sys.maxsize
    
    # 최단 거리 테이블을 모두 무한으로 초기화
    distances = [INF] * (V + 1)
    distances[start] = 0
    
    # 우선순위 큐 (힙) 초기화, (거리, 정점) 튜플로 저장
    priority_queue = [(0, start)]
    
    while priority_queue:
        current_distance, current_vertex = heapq.heappop(priority_queue)
        
        # 이미 처리된 정점이면 무시
        if current_distance > distances[current_vertex]:
            continue
        
        # 현재 정점과 연결된 다른 정점들을 확인
        for neighbor, weight in graph[current_vertex]:
            distance = current_distance + weight
            
            # 더 짧은 경로를 발견하면 갱신
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))
    
    return distances

def main():
    input = sys.stdin.read
    data = input().split()
    
    V = int(data[0])
    E = int(data[1])
    K = int(data[2])
    
    graph = [[] for _ in range(V + 1)]
    
    index = 3
    for _ in range(E):
        u = int(data[index])
        v = int(data[index + 1])
        w = int(data[index + 2])
        graph[u].append((v, w))
        index += 3
    
    distances = dijkstra(V, graph, K)
    
    for i in range(1, V + 1):
        if distances[i] == sys.maxsize:
            print("INF")
        else:
            print(distances[i])

if __name__ == "__main__":
    main()
```
