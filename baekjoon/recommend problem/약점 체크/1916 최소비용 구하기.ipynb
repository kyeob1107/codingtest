```python
# 뤼튼이 알려준 다익스트라 알고리즘 우선순위 큐(힙)이용하는 방식 -> 왜 자꾸 2%대에서 틀리는지 모르겠음
# 이유 찾음 중복 노선 있을 경우 무조건 나중에 나온 노선의 값으로 계산해서 그럼, 그런데 아직 heapq이해 못해서 아직 수정못함
import heapq
import sys
# input = sys.stdin.readline
input = open("../../input.txt", "rt").readline
N = int(input().strip()) # 도시갯수
M = int(input().strip()) # 버스갯수

graph = {node: dict() for node in range(N + 1)}
for _ in range(M):
    a, b, c = map(int, input().split())
    graph[a][b] = c

start, end = map(int, input().split())

def dijkstra2(graph, start):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    queue = [(0, start)]

    while queue:
        current_distance, current_node = heapq.heappop(queue)

        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(queue, (distance, neighbor))

    return distances


print(dijkstra2(graph, start)[end])
```

```python
# 레퍼런스
from collections import deque
import sys


input = sys.stdin.readline


N, M = int(input()), int(input())
arr = [list(map(int, input().strip().split())) for _ in range(M)]
graph = [[-1] * (N+1) for _ in range(N+1)]
for i, j, cost in arr:
    if graph[i][j] == -1 or graph[i][j] > cost:
        graph[i][j] = cost

visit = [False] * (N+1)
distance = [0] * (N+1)
q = deque()
start, end = list(map(int, input().strip().split()))
q.append(start)
visit[start] = True
while q:
    now = q.popleft()
    for i, cost in enumerate(graph[now]):
        if cost > -1 and (not visit[i] or (visit[i] and distance[i] > distance[now] + cost)):
            q.append(i)
            visit[i] = True
            distance[i] = distance[now] + cost

print(distance[end])
```

```python
# 2등 코드
import heapq
import os


def main():
    nums = map(int, os.read(0, os.stat(0).st_size).split())
    N = next(nums)
    M = next(nums)
    G = [dict() for _ in range(N + 1)]
    for _ in range(M):
        u = next(nums)
        v = next(nums)
        w = next(nums)
        pw = G[u].get(v, 1 << 29)
        if w < pw:
            G[u][v] = w
    s = next(nums)
    t = next(nums)
    dist = [1 << 29] * (N + 1)
    dist[s] = 0
    p = 2 ** -20
    queue = [0 + s * p]
    while queue:
        i = heapq.heappop(queue)
        du = int(i // 1)
        u = int(i % 1 / p)
        if u == t:
            break
        if du > dist[u]:
            continue
        for v, w in G[u].items():
            dv = du + w
            if dv < dist[v]:
                dist[v] = dv
                heapq.heappush(queue, dv + v * p)
    os.write(1, str(dist[t]).encode())
    os._exit(0)


main()
```
