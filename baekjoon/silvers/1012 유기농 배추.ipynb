```python
# bfs 방식 시간초과
import sys
from collections import deque
# input = sys.stdin.readline
input = open("input.txt", "rt").readline

def bfs(i, j):
    move = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    queue = deque([(j, i)])
    while queue:
        x, y = queue.popleft()
        visited[y][x] = True
        for dx, dy in move:
            nx = x + dx
            ny = y + dy
            if 0 <= nx < M and 0 <= ny < N and farm[ny][nx] and not visited[ny][nx]:
                queue.append((nx, ny))
        

T = int(input())
ans = []
for _ in range(T):
    M, N, K = map(int, input().split())
    farm = [[False] * M for _ in range(N)]
    visited = [[False] * M for _ in range(N)]
    cnt = 0
    for _ in range(K):
        X, Y = map(int, input().split())
        farm[Y][X] = True
    
    for i in range(N):
        for j in range(M):
            if farm[i][j] and not visited[i][j]:
                bfs(i, j)
                cnt += 1
    ans.append(cnt)

for a in ans:
    print(a)

```

```python
import sys
from collections import deque
# input = sys.stdin.readline
input = open("input.txt", "rt").readline

def bfs(start_x, start_y):
    move = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    queue = deque([(start_x, start_y)])
    positions.remove((start_x, start_y))
    
    while queue:
        x, y = queue.popleft()
        for dx, dy in move:
            nx, ny = x + dx, y + dy
            if (nx, ny) in positions:
                positions.remove((nx, ny))
                queue.append((nx, ny))

T = int(input().strip())
results = []

for _ in range(T):
    M, N, K = map(int, input().strip().split())
    positions = set()
    
    for _ in range(K):
        X, Y = map(int, input().strip().split())
        positions.add((X, Y))

    cnt = 0
    while positions:  # positions이 비어질 때까지 반복
        px, py = next(iter(positions))  # positions의 임의의 요소 하나 가져오기
        bfs(px, py)
        cnt += 1
    
    results.append(cnt)

for result in results:
    print(result)
```
