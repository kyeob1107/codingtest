```python
import sys

input = open("회의실 예약.txt","rt").readline
# N, M = map(int, sys.stdin.readline().split())
N, M = map(int, input().split())

room_info = dict()

def info(name):
    switch = False
    able_list = []
    for i, check in enumerate(room_info[name]):
        if check:
            if not switch:
                s_t = '0' + str(i + 9) if i + 9 < 10 else str(i + 9)
            switch = True
            
        else:
            if switch:
                # able_list.append([s_t, 
                #                   '0' + str(i + 9) if i + 9 < 10 else str(i + 9)])
                able_list.append([s_t, str(i + 9)])
            switch = False
        
        if switch and (i == len(room_info[name]) - 1):
            able_list.append([s_t, '18'])

    print(f'Room {name}:')
    if able_list:
        print(f'{len(able_list)} available:')
        for time in able_list:
            print(f'{time[0]}-{time[1]}')

    else:
        print('Not available')
            

for _ in range(N):
    name_temp = input().strip()
    room_info[name_temp]=[True] * 9
    
for _ in range(M):
    name, start, end = input().split()
    for n in range(int(start), int(end)):
        room_info[name][n - 9] = False

for j, room in enumerate(sorted(room_info.keys())):
    info(room)
    if j < len(room_info.keys()) - 1:
        print("-----")
```
